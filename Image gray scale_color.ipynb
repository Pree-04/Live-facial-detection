```python
# importing libraries.
import matplotlib.pyplot as plt
from PIL import Image

# storing image path
fname = r'king kong.jpg'
 
# opening image using pil 
image = Image.open(fname).convert("L") # L-Luminance,
plt.imshow(image,cmap='gray') # mapping image to gray scale
plt.show()

image = Image.open(fname).convert("P") # P-Palette,RGB
plt.imshow(image)
plt.show()
```

## Convert 2d Array Into Grayscale Image.

```python
from matplotlib import pyplot as plt
import numpy as np

plt.figure(figsize=(3,3))
plt.axis(False)


# Set the figure size
data = np.random.rand(5, 5)# Random data points
print(data)

plt.imshow(data, cmap='gray') # Plot the data using imshow with gray colormap
plt.show() # Display the plot
```

```python
import numpy, cv2

b = numpy.zeros([5,4,3])

b[:,:,0] = numpy.random.rand(5,4)*255
b[:,:,1] = numpy.random.rand(5,4)*255
b[:,:,2] = numpy.random.rand(5,4)*255
cv2.imwrite('color_img.jpg', b)


d = cv2.imread('color_img.jpg', 1)
d = cv2.cvtColor(d, cv2.COLOR_BGR2RGB)

plt.imshow(d)
plt.show()
```

```python
import numpy, cv2

b = numpy.zeros([200,200,3])

b[:,:,0] = numpy.ones([200,200])*0
b[:,:,1] = numpy.ones([200,200])*0
b[:,:,2] = numpy.ones([200,200])*255
cv2.imwrite('color_img.jpg', b)


d = cv2.imread('color_img.jpg', 1)
d = cv2.cvtColor(d, cv2.COLOR_BGR2RGB)

plt.imshow(d)
plt.show()
```
